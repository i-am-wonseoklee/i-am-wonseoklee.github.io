```python
import numpy as np
import matplotlib.pyplot as plt

def gen_kernel1() -> np.ndarray:
    return np.array([-1, 1])

def gen_kernel2() -> np.ndarray:
    return np.array([0, -1, 1, 0])

def gen_kernel3() -> np.ndarray:
    return np.array([-1, 2])

def gen_signal() -> np.ndarray:
    signal = np.zeros(31)
    signal[:] = -1
    signal[10:20] = 1
    return signal

def convolve(signal: np.ndarray, kernel: np.ndarray) -> np.ndarray:
    return np.convolve(signal, kernel, 'same')
```

```python
signal = gen_signal()

fig, ax = plt.subplots(3, 3)

# Kernel 1.
kernel = gen_kernel1()
result = convolve(signal, kernel)
ax[0, 0].plot(range(len(kernel)), kernel)
ax[0, 1].plot(range(len(signal)), signal)
ax[0, 2].plot(range(len(signal)), signal)
ax[0, 2].plot(range(len(signal)), result)

# Kernel 2.
kernel = gen_kernel2()
result = convolve(signal, kernel)
ax[1, 0].plot(range(len(kernel)), kernel)
ax[1, 1].plot(range(len(signal)), signal)
ax[1, 2].plot(range(len(signal)), signal)
ax[1, 2].plot(range(len(signal)), result)

# Kernel 3.
kernel = gen_kernel3()
result = convolve(signal, kernel)
ax[2, 0].plot(range(len(kernel)), kernel)
ax[2, 1].plot(range(len(signal)), signal)
ax[2, 2].plot(range(len(signal)), signal)
ax[2, 2].plot(range(len(signal)), result)

plt.show()
```
