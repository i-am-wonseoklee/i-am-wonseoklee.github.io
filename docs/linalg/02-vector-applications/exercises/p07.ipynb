```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gen_kernel() -> np.ndarray:
    kernel = np.array([0, 0.1, 0.3, 0.8, -1, 0.8, 0.3, 0.1, 0])
    return kernel / np.sum(kernel)

def gen_signal() -> np.ndarray:
    return np.random.randn(200)
```

```python
kernel = gen_kernel()
signal = gen_signal()
result = np.convolve(signal, kernel, mode='same')
```

```python
fig, ax = plt.subplots(1, 3)
ax[0].plot(range(len(kernel)), kernel)
ax[1].set_ylim(-10, 10)
ax[1].plot(range(len(signal)), signal)
ax[2].set_ylim(-10, 10)
ax[2].plot(range(len(result)), result)
plt.show()
```
