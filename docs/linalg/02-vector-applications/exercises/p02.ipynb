```python
import numpy as np
import matplotlib.pyplot as plt

from typing import List

def do_mean_centering(v: np.ndarray) -> np.ndarray:
    return v - np.mean(v)

def get_cosine_similarity(v1: np.ndarray, v2: np.ndarray) -> float:
    return np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))

def get_pearson_correlation(v1: np.ndarray, v2: np.ndarray) -> float:
    return get_cosine_similarity(do_mean_centering(v1), do_mean_centering(v2))

def step(offset: float) -> List[float]:
    v1 = np.linspace(0, 3, 4, endpoint=True)
    v2 = v1 + offset
    return get_cosine_similarity(v1, v2), get_pearson_correlation(v1, v2)
```

```python
fig, ax = plt.subplots(1, 1)
ax.set_title("Cosine similarity vs. Pearson correlation")
ax.set_xlabel("offset")
ax.set_xlim(-51, 51)
ax.set_ylabel("coefficients")
ax.set_ylim(-2, 2)

x = np.linspace(-50, 50, 101, endpoint=True)
y = np.array([step(offset) for offset in x])

ax.scatter(x, y[:, 0], label="cosine similarity")
ax.scatter(x, y[:, 1], label="pearson correlation")

plt.legend()
plt.show()
```
