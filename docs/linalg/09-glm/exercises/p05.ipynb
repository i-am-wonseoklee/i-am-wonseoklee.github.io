```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.random.randn(6, 6)
E = np.eye(6)
```

```python
INV_1 = np.zeros_like(X)
XTX_INV_XT = np.linalg.inv(X.T @ X) @ X.T
for c in range(X.shape[1]):
    INV_1[:, c] = np.linalg.inv(X.T @ X) @ X.T @ E[:, c]
```

```python
INV_2 = np.linalg.inv(X.T @ X) @ X.T @ E
```

```python
INV_3 = np.linalg.inv(X)
```

```python
fig, axs = plt.subplots(1, 3)

axs[0].set_title("Column-wise least square")
axs[0].imshow(np.round(INV_1 @ X, 6))

axs[1].set_title("Matrix-wise least square")
axs[1].imshow(np.round(INV_2 @ X, 6))

axs[2].set_title("inv()")
axs[2].imshow(np.round(INV_3 @ X, 6))

fig.tight_layout()
plt.show()
```
