```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
NCOURSES_1 = np.array([13,4,12,3,14,13,12,9,11,7,13,11,9,2,5,7,10,0,9,7]).reshape(-1, 1)
HAPPINESS_1 = np.array([70,25,54,21,80,68,84,62,57,40,60,64,45,38,51,52,58,21,75,70]).reshape(-1, 1)
```

```python
NCOURSES_2 = np.array([13,4,12,3,14,13,12,9,11,7,13,11,9,2,5,7,10,0,9,7]).reshape(-1, 1)
HAPPINESS_2 = np.array([170,25,54,21,80,68,84,62,57,40,60,64,45,38,51,52,58,21,75,70]).reshape(-1, 1)
```

```python
NCOURSES_3 = np.array([13,4,12,3,14,13,12,9,11,7,13,11,9,2,5,7,10,0,9,7]).reshape(-1, 1)
HAPPINESS_3 = np.array([70,25,54,21,80,68,84,62,57,40,60,64,45,38,51,52,58,21,75,170]).reshape(-1, 1)
```

```python
X_1 = np.hstack((np.ones_like(NCOURSES_1), NCOURSES_1))
X_2 = np.hstack((np.ones_like(NCOURSES_2), NCOURSES_2))
X_3 = np.hstack((np.ones_like(NCOURSES_3), NCOURSES_3))
beta_1 = np.linalg.inv(X_1.T @ X_1) @ X_1.T @ HAPPINESS_1
beta_2 = np.linalg.inv(X_2.T @ X_2) @ X_2.T @ HAPPINESS_2
beta_3 = np.linalg.inv(X_3.T @ X_3) @ X_3.T @ HAPPINESS_3
sse_1 = np.sum((X_1 @ beta_1 - HAPPINESS_1)**2)
sse_2 = np.sum((X_2 @ beta_2 - HAPPINESS_2)**2)
sse_3 = np.sum((X_3 @ beta_3 - HAPPINESS_3)**2)
```

```python
fig, axs = plt.subplots(1, 3)
axs[0].set_title(f"SSE: {sse_1:.2f}")
axs[0].scatter(NCOURSES_1, HAPPINESS_1, label="Data")
axs[0].scatter(NCOURSES_1, X_1 @ beta_1, label="Prediction")
axs[1].set_title(f"SSE: {sse_2:.2f}")
axs[1].scatter(NCOURSES_2, HAPPINESS_2, label="Data")
axs[1].scatter(NCOURSES_2, X_2 @ beta_2, label="Prediction")
axs[2].set_title(f"SSE: {sse_3:.2f}")
axs[2].scatter(NCOURSES_3, HAPPINESS_3, label="Data")
axs[2].scatter(NCOURSES_3, X_3 @ beta_3, label="Prediction")
fig.tight_layout()
plt.show()
```
