```python
import matplotlib.pyplot as plt
import numpy as np
import scipy
```

```python
np_det = []
lu_det = []
plu_det = []
for i in range(100):
    A = np.random.rand(50, 50)
    np_det.append(np.linalg.det(A))
    P, L, U = scipy.linalg.lu(A)
    lu_det.append(np.prod(np.diag(L)) * np.prod(np.diag(U)))
    plu_det.append(np.linalg.det(P) * lu_det[-1])
```

```python
fig, axs = plt.subplots(1)

axs.scatter(range(100), np_det, label="np.linalg.det")
axs.plot(range(100), lu_det, label="det(L) * det(U)")
axs.plot(range(100), plu_det, label="det(P) * det(L) * det(U)")

plt.legend()
plt.show()
```
