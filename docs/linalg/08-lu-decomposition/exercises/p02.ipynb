```python
import matplotlib.pyplot as plt
import numpy as np
import scipy
```

```python
# Make 6x8 matrix w/ rank 3.
A = np.random.rand(6, 3)
B = np.random.rand(3, 8)
C = A @ B
rank_C = np.linalg.matrix_rank(C)
```

```python
# LU decomposition.
_, L, U = scipy.linalg.lu(C)
rank_L = np.linalg.matrix_rank(L)
rank_U = np.linalg.matrix_rank(U)
```

```python
fig, axs = plt.subplots(1, 3)

axs[0].set_title(f"C (rank: {rank_C})")
axs[0].imshow(C)

axs[1].set_title(f"L (rank: {rank_L})")
axs[1].imshow(L)

axs[2].set_title(f"U (rank: {rank_U})")
axs[2].imshow(U)

plt.show()
```
