```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
CNUM = 42
```

```python
R0 = np.random.randn(60, 60)
R1 = np.random.randn(30, 30)
U, _ = np.linalg.qr(R0)
VT, _ = np.linalg.qr(R1)
S = np.zeros((60, 30))
np.fill_diagonal(S, np.linspace(1, CNUM, 30, endpoint=True))
A = U @ S @ VT
```

```python
fig, axs = plt.subplots(1, 4)

axs[0].set_title(f"A cond: {np.linalg.cond(A)}")
axs[0].imshow(A)

axs[1].set_title("U")
axs[1].imshow(U)

axs[2].set_title("Sigma")
axs[2].imshow(S)

axs[3].set_title("VT")
axs[3].imshow(VT)

fig.tight_layout()
plt.show()
```
