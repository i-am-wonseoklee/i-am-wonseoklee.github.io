```python
import matplotlib.pyplot as plt
import numpy as np
import scipy
```

```python
# A.
M = np.random.randn(30, 40)
x, y = np.meshgrid(range(-3, 4), range(-3, 4))
kernel = np.exp(-(x**2 + y**2) / 9)
G = scipy.signal.convolve2d(M, kernel, mode="same")
U, s, Vt = np.linalg.svd(G)
S = np.zeros((30, 40))
np.fill_diagonal(S, s)
```

```python
# Visualize A.
fig, axs = plt.subplots(1, 4)
axs[0].set_title("G")
axs[0].imshow(G)
axs[1].set_title("U")
axs[1].imshow(U)
axs[2].set_title("Sigma")
axs[2].imshow(S)
axs[3].set_title("Vt")
axs[3].imshow(Vt)
fig.tight_layout()
plt.show()
```

```python
# B
denom = np.sum(S.diagonal())
var = (100 * S.diagonal()) / denom
```

```python
# Visualize B.
fig, ax = plt.subplots()
ax.plot(range(len(S.diagonal())), var, "o")
ax.set_xlabel("nth singular value")
ax.set_ylabel("variance %")
fig.tight_layout()
plt.show()
```

```python
# C.
L0 = s[0] * np.outer(U[:, 0], Vt[0, :])
L1 = s[1] * np.outer(U[:, 1], Vt[1, :])
L2 = s[2] * np.outer(U[:, 2], Vt[2, :])
L3 = s[3] * np.outer(U[:, 3], Vt[3, :])
L00 = L0
L01 = L0 + L1
L02 = L0 + L1 + L2
L03 = L0 + L1 + L2 + L3
```

```python
# Visualize C.
fig, axs = plt.subplots(2, 4)

axs[0][0].set_title("L0")
axs[0][0].imshow(L0)
axs[0][1].set_title("L1")
axs[0][1].imshow(L1)
axs[0][2].set_title("L2")
axs[0][2].imshow(L2)
axs[0][3].set_title("L3")
axs[0][3].imshow(L3)
axs[1][0].set_title("L00")
axs[1][0].imshow(L00)
axs[1][1].set_title("L01")
axs[1][1].imshow(L01)
axs[1][2].set_title("L02")
axs[1][2].imshow(L02)
axs[1][3].set_title("L03")
axs[1][3].imshow(L03)

fig.tight_layout()
plt.show()
```
