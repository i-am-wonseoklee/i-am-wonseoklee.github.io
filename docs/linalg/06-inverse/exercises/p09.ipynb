```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def get_hilbert(k):
    n = np.arange(0, k).reshape(1, -1)
    return 1 / (n.T + 1 + n)
```

```python
H = get_hilbert(5)
H_inv = np.linalg.inv(H)
I = H_inv @ H
```

```python
fig, axes = plt.subplots(1, 3)

axes[0].set_title("Hilbert matrix")
h0 = axes[0].imshow(H, cmap="gray")
fig.colorbar(h0, ax=axes[0], fraction=.045)

axes[1].set_title("Inverse of Hilbert matrix")
h1 = axes[1].imshow(H_inv, cmap="gray")
fig.colorbar(h1, ax=axes[1], fraction=.045)

axes[2].set_title("Product")
h2 = axes[2].imshow(I, cmap="gray")
fig.colorbar(h2, ax=axes[2], fraction=.045)

fig.tight_layout()
plt.show()
```
