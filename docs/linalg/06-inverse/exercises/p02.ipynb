```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def get_minor(m):
    ret = np.zeros_like(m)
    for r in range(ret.shape[0]):
        for c in range(ret.shape[1]):
            rmask = [True] * ret.shape[0]
            rmask[r] = False
            cmask = [True] * ret.shape[1]
            cmask[c] = False
            ret[r, c] = np.linalg.det(m[rmask, :][:, cmask])
    return ret
```

```python
def get_grid(m):
    ret = np.zeros_like(m)
    for r in range(ret.shape[0]):
        for c in range(ret.shape[1]):
            ret[r, c] = (-1)**(r + c)
    return ret
```

```python
def get_cofactor(m):
    return get_minor(m) * get_grid(m)
```

```python
def get_adjugate(m):
    return get_cofactor(m).T / np.linalg.det(m)
```

```python
M = np.random.randn(4, 4)
minor = get_minor(M)
grid = get_grid(M)
cofactor = get_cofactor(M)
adjugate = get_adjugate(M)
inv = np.linalg.inv(M)
eye = M @ inv
```

```python
fig, axes = plt.subplots(2, 3)

axes[0, 0].set_title("minor")
axes[0, 0].imshow(minor, cmap="gray")

axes[0, 1].set_title("grid")
axes[0, 1].imshow(grid, cmap="gray")

axes[0, 2].set_title("cofactor")
axes[0, 2].imshow(cofactor, cmap="gray")

axes[1, 0].set_title("adjugate")
axes[1, 0].imshow(adjugate, cmap="gray")

axes[1, 1].set_title("np.linalg.inv")
axes[1, 1].imshow(inv, cmap="gray")

axes[1, 2].set_title("A @ inv")
axes[1, 2].imshow(eye, cmap="gray")

fig.tight_layout()
plt.show()
```
