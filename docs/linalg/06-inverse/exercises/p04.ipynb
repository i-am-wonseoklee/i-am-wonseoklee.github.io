```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
M = np.random.randn(4, 40)
M_inv = M.T @ np.linalg.inv(M @ M.T)
```

```python
fig, axes = plt.subplots(2, 2)

axes[0, 0].set_title("M")
axes[0, 0].imshow(M, cmap="gray")

axes[0, 1].set_title("M_inv (right inverse)")
axes[0, 1].imshow(M_inv, cmap="gray")

axes[1, 0].set_title("M @ M_inv")
axes[1, 0].imshow(M @ M_inv, cmap="gray")

axes[1, 1].set_title("M_inv @ M")
axes[1, 1].imshow(M_inv @ M, cmap="gray")

fig.tight_layout()
plt.show()
```
