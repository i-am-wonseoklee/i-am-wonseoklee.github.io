```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
T = np.array([[1.0, 0.5], [0.0, 0.5]])
theta = np.linspace(0, 2 * np.pi, 30)
org = np.vstack((np.cos(theta), np.sin(theta)))
dst = T @ org
rev = np.linalg.inv(T) @ dst
```

```python
fig, ax = plt.subplots()
ax.scatter(org[0, :], org[1, :], label="org")
ax.scatter(dst[0, :], dst[1, :], label="dst")
ax.scatter(rev[0, :], rev[1, :], label="rev", marker='x')
ax.set_aspect("equal")
plt.tight_layout()
plt.show()
```
