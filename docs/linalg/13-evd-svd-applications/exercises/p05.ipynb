```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
class0 = np.random.normal(0, 1, (200, 2))
class0[:, 0] += class0[:, 1]
class0 += np.array([2, -1])

class1 = np.random.normal(0, 1, (200, 2))
class1[:, 0] += class1[:, 1]

data = np.vstack((class0, class1))
label = np.vstack((np.zeros((200, 1)), np.ones((200, 1))))
```

```python
ax = sns.jointplot(x=data[:, 0], y=data[:, 1], hue=label[:, 0])
ax.ax_joint.set_xlabel("x")
ax.ax_joint.set_ylabel("y")
ax.plot_joint(sns.kdeplot)
plt.show()
```
