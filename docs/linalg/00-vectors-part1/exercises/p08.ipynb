```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 8))
ax.set_xlim(0, 5)
ax.set_ylim(0, 5)
ax.set_xlabel('x')
ax.set_ylabel('y')

a = np.array([1, 4]).reshape(-1, 1)
b = np.array([4, 1]).reshape(-1, 1)

ax.arrow(0, 0, *a[:, 0], head_width=0.2, head_length=0.2, label="a", color="r")
ax.arrow(0, 0, *b[:, 0], head_width=0.2, head_length=0.2, label="b", color="b")

beta = np.dot(b.ravel(), a.ravel()) / np.dot(a.ravel(), a.ravel())
foot = beta * a
ax.arrow(*foot[:, 0], *(b - foot)[:, 0], label="b - foot", color="g")

ax.legend()
plt.show()
```
