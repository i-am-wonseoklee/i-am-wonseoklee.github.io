```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
year       = np.array([1534, 1737, 1803, 1928, 1960, 1975, 1987, 2023, 2057, 2100 ])
doubleTime = np.array([ 697,  594,  260,  125,   76,   47,   37,   48,   70,  109 ])
```

```python
pred = np.zeros((10, year.shape[0]))
for deg in range(10):
    beta = np.polyfit(year, doubleTime, deg)
    pred[deg] = np.polyval(beta, year)
```

```python
fig, axs = plt.subplots(5, 2)
for deg in range(10):
    ax = axs[deg // 2, deg % 2]
    ax.set_title(f"Degree: {deg}")
    ax.plot(year, doubleTime, "bo-")
    ax.plot(year, pred[deg], "rx-")
fig.tight_layout()
plt.show()
```
