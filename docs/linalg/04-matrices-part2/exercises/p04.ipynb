```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mat = np.random.randn(10, 10)
norm = np.sqrt(np.sum(mat**2))
```

```python
rate = np.empty(30)
dist = np.empty(30)
corr = np.empty(30)

scales = np.linspace(0, 1, 30)
for idx, s in enumerate(scales):
    new_mat = mat + s * norm * np.eye(10)
    rate[idx] = np.sqrt(np.sum(new_mat**2)) / norm
    dist[idx] = np.sqrt(np.sum((new_mat - mat)**2))
    corr[idx] = np.corrcoef(mat.flatten(), new_mat.flatten())[0, 1]

fig, ax = plt.subplots(1, 3)

ax[0].set_title("rate [%]")
ax[0].plot(scales, rate * 100)

ax[1].set_title("distance")
ax[1].plot(scales, dist)

ax[2].set_title("correlation")
ax[2].plot(scales, corr)

plt.show()
```
