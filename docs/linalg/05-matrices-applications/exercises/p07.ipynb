```python
import matplotlib.pyplot as plt
import numpy as np

from scipy.signal import convolve2d
from skimage import color, io
```

```python
IMAGE_URL = "https://upload.wikimedia.org/wikipedia/commons/6/61/De_nieuwe_vleugel_van_het_Stedelijk_Museum_Amsterdam.jpg"
image = io.imread(IMAGE_URL)
image = color.rgb2gray(image)
```

```python
vkernel = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])
hkernel = np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]])
```

```python
vimage = convolve2d(image, vkernel)
himage = convolve2d(image, hkernel)
```

```python
fig, ax = plt.subplots(1, 2)
ax[0].set_title("V")
ax[0].imshow(vimage, cmap="gray", vmin=0, vmax=.01)
ax[1].set_title("H")
ax[1].imshow(himage, cmap="gray", vmin=0, vmax=.01)
plt.show()
```
