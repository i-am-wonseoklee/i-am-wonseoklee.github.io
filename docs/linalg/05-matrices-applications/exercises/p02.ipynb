```python
data = pd.read_csv(DATA_URL, sep=",", header=None)
data.columns = DATA_COL_NAMES
data = data._get_numeric_data()
data = data.drop(["state", "fold"], axis=1).values
```

```python
def my_cor(data):
    data_var = data - np.mean(data, axis=0)
    cov = data_var.T @ data_var / (data.shape[0] - 1)
    sqrt_inv = np.diag(1 / np.sqrt(np.diag(cov)))
    return sqrt_inv @ cov @ sqrt_inv
```

```python
my_cor = my_cor(data)
np_cor = np.corrcoef(data.T)
diff = my_cor - np_cor

fig, axes = plt.subplots(1, 3)

axes[0].set_title("my cor")
h0 = axes[0].imshow(my_cor, vmin=np.min(my_cor), vmax=np.max(my_cor), cmap="gray")

axes[1].set_title("np_cor")
h1 = axes[1].imshow(np_cor, vmin=np.min(np_cor), vmax=np.max(np_cor), cmap="gray")

axes[2].set_title("diff")
h2 = axes[2].imshow(diff, vmin=np.min(diff), vmax=np.max(diff), cmap="gray")

fig.colorbar(h0, ax=axes[0], fraction=0.045)
fig.colorbar(h1, ax=axes[1], fraction=0.045)
fig.colorbar(h2, ax=axes[2], fraction=0.045)
plt.tight_layout()
plt.show()
```
