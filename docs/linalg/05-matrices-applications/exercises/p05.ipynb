```python
import matplotlib.pyplot as plt
import numpy as np

from scipy.signal import convolve2d
from skimage import io, color
```

```python
IMAGE_URL = "https://upload.wikimedia.org/wikipedia/commons/6/61/De_nieuwe_vleugel_van_het_Stedelijk_Museum_Amsterdam.jpg"
image = io.imread(IMAGE_URL)
```

```python
Y, X = np.meshgrid(np.linspace(-3, 3, 7, endpoint=True), np.linspace(-3, 3, 7, endpoint=True))
kernel = np.exp(-(X**2 + Y**2) / 7)
kernel /= np.sum(kernel)
```

```python
smooth_image = np.zeros_like(image)
for ch in range(image.shape[2]):
    smooth_image[:, :, ch] = convolve2d(image[:, :, ch], kernel, mode="same")
```

```python
fig, ax = plt.subplots(1, 2)

ax[0].set_title("org")
ax[0].imshow(image)

ax[1].set_title("smooth")
ax[1].imshow(smooth_image)

plt.show()
```
