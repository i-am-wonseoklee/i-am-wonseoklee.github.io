```python
import numpy as np
import plotly.graph_objects as go

bases = np.array([[3., 0.], [5., 2.], [1., 2.]])
scalars = np.random.uniform(-4, 4, (2, 100))
results = bases @ scalars

fig = go.Figure(data=[go.Scatter3d(x=results[0,:], y=results[1,:], z=results[2,:], mode='markers')])
fig.show()
```
