```python
import numpy as np
import matplotlib.pyplot as plt

m = np.arange(100).reshape((10, 10))
m_ul = m[0:5, 0:5]
m_ur = m[0:5, 5:]
m_ll = m[5:, 0:5]
m_lr = m[5:, 5:]

s = np.vstack((np.hstack((m_lr, m_ll)), np.hstack((m_ur, m_ul))))

_, ax = plt.subplots(1, 2)

ax[0].set_title("original")
ax[1].set_title("shifted")

ax[0].imshow(m, cmap="inferno", origin="upper", vmin=0, vmax=99)
ax[1].imshow(s, cmap="inferno", origin="upper", vmin=0, vmax=99)

plt.show()
```
