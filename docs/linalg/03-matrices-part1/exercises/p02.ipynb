```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 2)

m = np.arange(100).reshape((10, 10))
s = m[0:5, 0:5]

ax[0].set_title("original")
ax[0].imshow(m, cmap="inferno", origin="upper", vmin=np.min(m), vmax=np.max(m))
ax[1].set_title("submatrix")
ax[1].imshow(s, cmap="inferno", origin="upper", vmin=np.min(m), vmax=np.max(m))

plt.show()
```
