```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
M = np.random.randn(2, 2)
M = M.T @ M
```

```python
_, evecs = np.linalg.eig(M)
w1 = evecs[0, :].reshape(-1, 1)
w2 = evecs[1, :].reshape(-1, 1)
```

```python
Mw1 = M @ w1
Mw2 = M @ w2
```

```python
org = np.array([0, 0]).reshape(-1, 1)
```

```python
w1 = np.hstack((org, w1))
w2 = np.hstack((org, w2))
Mw1 = np.hstack((org, Mw1))
Mw2 = np.hstack((org, Mw2))
```

```python
fig, axs = plt.subplots()
axs.plot(w1[0, :], w1[1, :], "--", color="b", label="w1")
axs.plot(w2[0, :], w2[1, :], "--", color="r", label="w2")
axs.plot(Mw1[0, :], Mw1[1, :], color="b", label="Mw1")
axs.plot(Mw2[0, :], Mw2[1, :], color="r", label="Mw2")
plt.legend()
plt.show()
```
