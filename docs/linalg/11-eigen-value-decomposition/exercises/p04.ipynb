```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Rs = np.random.randn(123, 42, 42)
evalss = np.zeros((123, 42, 1)).astype(np.complex128) # type is important.
for i in range(Rs.shape[0]):
    R = Rs[i, :, :]
    evals, _ = np.linalg.eig(R)
    evalss[i] = evals.reshape(-1, 1)
```

```python
evalss = evalss.flatten() / 42**0.5
reals = [n.real for n in evalss]
imags = [n.imag for n in evalss]
```

```python
fig, ax = plt.subplots()
ax.set_xlabel("real part")
ax.set_ylabel("imag part")
ax.scatter(reals, imags, marker="o")
plt.show()
```
