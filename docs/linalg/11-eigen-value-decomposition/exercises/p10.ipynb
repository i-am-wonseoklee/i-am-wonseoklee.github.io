```python
import matplotlib.pyplot as plt
import numpy as np
import scipy
```

```python
R0 = np.random.randn(5, 5)
R1 = np.random.randn(5, 5)
A = R0.T @ R0
B = R1.T @ R1
```

```python
evals, evecs = scipy.linalg.eigh(A, B)
```

```python
fig, axs = plt.subplots(1, 2)
axs[0].set_title("V.T @ V")
axs[0].imshow(evecs @ evecs.T)
axs[1].set_title("V.T @ B @ V")
axs[1].imshow(evecs.T @ B @ evecs)
plt.show()
```
