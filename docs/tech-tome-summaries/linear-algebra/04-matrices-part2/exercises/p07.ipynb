```python
import numpy as np
import matplotlib.pyplot as plt

def gen_rank_r_matrix(nrow, rank):
    return np.random.randn(nrow, rank) @ np.random.randn(rank, nrow)

rank_add = np.empty((14, 14))
rank_mul = np.empty((14, 14))

for ia, ra in enumerate(range(2, 16)):
    ma = gen_rank_r_matrix(20, ra)
    for ib, rb in enumerate(range(2, 16)):
        mb = gen_rank_r_matrix(20, rb)
        rank_add[ia, ib] = np.linalg.matrix_rank(ma * mb)
        rank_mul[ia, ib] = np.linalg.matrix_rank(ma @ mb)

fig, ax = plt.subplots(1, 2)
ax[0].set_ylabel("rank of A + B")
ax[0].set_xlabel("rank of A")
a0 = ax[0].imshow(rank_add, cmap="inferno", origin="lower", vmin=2, vmax=16)
fig.colorbar(a0, ax=ax[0], fraction=.045)
ax[1].set_ylabel("rank of A @ B")
ax[1].set_xlabel("rank of A")
a1 = ax[1].imshow(rank_mul, cmap="inferno", origin="lower", vmin=2, vmax=16)
fig.colorbar(a1, ax=ax[1], fraction=.045)
plt.tight_layout()
plt.show()
        
```
