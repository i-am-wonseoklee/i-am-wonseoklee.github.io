```python
import numpy as np

def is_v_in_col_space(A, v):
    aug = np.hstack((A, v))
    if np.linalg.matrix_rank(A) == np.linalg.matrix_rank(aug):
        return True
    else:
        print("v is not in A.")
        return False


cnt = 0
for _ in range(100):
    A = np.random.randn(4, 3)
    v = np.random.randn(4, 1)
    cnt += 1 if is_v_in_col_space(A, v) else 0
print(cnt)

cnt = 0
for _ in range(100):
    A = np.random.randn(4, 4)
    v = np.random.randn(4, 1)
    cnt += 1 if is_v_in_col_space(A, v) else 0
print(cnt)
```
