```python
import numpy as np
import matplotlib.pyplot as plt

def get_frobenius_norm(m):
    return np.sqrt(np.trace(m @ m.T))

fs = np.empty((40, 10))

for i, s in enumerate(np.linspace(0, 50, 40)):
    for j in range(10):
        m = np.random.randn(10, 10) * s
        fs[i, j] = get_frobenius_norm(m)

fig, ax = plt.subplots()
ax.plot(np.linspace(0, 50, 40), np.mean(fs, axis=1))
plt.show()
        
```
