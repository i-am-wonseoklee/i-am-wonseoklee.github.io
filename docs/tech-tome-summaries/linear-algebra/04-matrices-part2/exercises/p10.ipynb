```python
import numpy as np
import matplotlib.pyplot as plt

ns = range(3, 31)
dets = np.empty(len(ns))
for i, n in enumerate(ns):
    sum = 0
    for _ in range(100):
        mat = np.random.randn(n, n)
        mat[:, 0] = mat[:, 1]
        sum += np.abs(np.linalg.det(mat))
    dets[i] = sum / 100.0

fig, ax = plt.subplots()
ax.plot(ns, np.log(dets))
plt.show()
    
        
```
