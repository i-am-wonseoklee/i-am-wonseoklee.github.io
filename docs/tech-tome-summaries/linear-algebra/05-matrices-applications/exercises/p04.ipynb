```python
import matplotlib.animation as ani
import matplotlib.pyplot as plt
import numpy as np
```

```python
def aframe(i, org, phi, h0, h1):
    T = np.array([[1.0 - phi[i] / 3.0, 0.0], [0.0, phi[i]]])
    
    dst = T @ org
    h1.set_xdata(dst[0, :])
    h1.set_ydata(dst[1, :])
    
    return (h0, h1)
```

```python
theta = np.linspace(-np.pi, np.pi, 100)
org = np.vstack((theta, np.sin(theta)))
```

```python
fig, ax = plt.subplots()
h0, = ax.plot(org[0, :], org[1, :], label="org")
h1, = ax.plot(org[0, :], org[1, :], label="dst")

phi = 1 - np.linspace(-1, 1, 100)
anim = ani.FuncAnimation(fig, aframe, frames=len(phi), fargs=(org, phi, h0, h1,), interval=50, repeat=True)
```
