```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 8))
ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_aspect("equal")

t = np.random.rand(2, 1)
r = np.random.rand(2, 1)
t = t / np.linalg.norm(t) * 3
r = r / np.linalg.norm(r) * 5

ax.arrow(0, 0, *t[:, 0], head_width=0.1, head_length=0.2, color="black", label="target")
ax.arrow(0, 0, *r[:, 0], head_width=0.1, head_length=0.2, color="red", label="reference")

alpha = np.dot(t.ravel(), r.ravel()) / np.dot(r.ravel(), r.ravel())
v_para = alpha * r
v_perp = t - v_para

ax.arrow(0, 0, *v_para[:, 0], head_width=0.1, head_length=0.2, color="blue", label="parallel")
ax.arrow(0, 0, *v_perp[:, 0], head_width=0.1, head_length=0.2, color="green", label="perpendicular")
ax.arrow(*v_para[:, 0], *v_perp[:, 0], color="blue")

alpha_wrong =  np.dot(r.ravel(), r.ravel()) / np.dot(t.ravel(), r.ravel())
v_para_wrong = alpha_wrong * r
v_perp_wrong = t - v_para_wrong

ax.arrow(0, 0, *v_para_wrong[:, 0], head_width=0.1, head_length=0.2, color="blue", label="parallel", alpha=0.5)
ax.arrow(0, 0, *v_perp_wrong[:, 0], head_width=0.1, head_length=0.2, color="green", label="perpendicular", alpha=0.5)
ax.arrow(*v_para_wrong[:, 0], *v_perp_wrong[:, 0], color="blue", alpha=0.5)

ax.legend()
plt.show()
```
