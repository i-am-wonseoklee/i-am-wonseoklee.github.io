```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
ax1, ax2 = axes

ax1.set_title("v + w")
ax1.set_xlim(-6, 6)
ax1.set_ylim(-6, 6)
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.set_aspect("equal")
ax1.grid()

v = np.array([[1, 2]]).T
w = np.array([[4, -6]]).T

ax1.arrow(0, 0, *v[:, 0], color="black", linewidth=2.0, label="v", head_width=0.2)
ax1.arrow(*v[:, 0], *w[:, 0], color="orange", linewidth=2.0, label="w", head_width=0.2)
ax1.arrow(0, 0, *(v + w)[:, 0], color="red", linewidth=2.0, label="v + w", head_width=0.2)
ax1.legend()

ax2.set_title("v - w")
ax2.set_xlim(-6, 6)
ax2.set_ylim(-6, 6)
ax2.set_xlabel("x")
ax2.set_ylabel("y")
ax2.set_aspect("equal")
ax2.grid()

v = np.array([[1, 2]]).T
w = np.array([[4, -6]]).T

ax2.arrow(0, 0, *v[:, 0], color="black", linewidth=2.0, label="v", head_width=0.2)
ax2.arrow(0, 0, *w[:, 0], color="orange", linewidth=2.0, label="w", head_width=0.2)
ax2.arrow(*w[:, 0], *(v - w)[:, 0], color="red", linewidth=2.0, label="v - w", head_width=0.2)
ax2.legend()

plt.show()
```
